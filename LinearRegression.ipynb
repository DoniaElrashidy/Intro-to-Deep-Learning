```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating dataset

```python
x = np.array([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = np.array([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

print(x, x.shape)
print(y, y.shape)

plt.plot(x, y, marker='o')
plt.show()

plt.scatter(x,y)
plt.show()
```

# Define 1D linear regression model

```python
def f(x, m, b):
  y = m * x + b
  return y
```

# Plot the function

```python
def plot(x, y, m, b):
 fig, ax = plt.subplots()
 ax.scatter(x, y)

 plt.xlim(0, 2)
 plt.ylim(0, 2)

 ax.set_xlabel('x')
 ax.set_ylabel('y')

 x_line = np.arange(0,2,0.01)
 y_line = f(x_line, m ,b)

 plt.plot(x_line, y_line, color='red')
 plt.show()


```

Initializing parameters

```python
m = -1.5
b = 1.25
plot(x, y, m, b)
```

# Calculate loss

```python
def compute_loss(x, y, m, b):
  loss = np.mean((y-f(x, m, b))**2)
  return loss
```

```python
loss = compute_loss(x, y, m, b)
print(f'{loss:3.3f}')
```

# Gradient Descent

```python
def gradient_descent(x, y, m, b, lr = 0.01, epochs = 100):
  for _ in range(epochs):
    dloss_dm = -2 * np.sum(x * (y - f(x, m, b)))
    dloss_db = -2 * np.sum(y - f(x, m, b))

    m -=  lr * dloss_dm
    b -=  lr * dloss_db

    loss = compute_loss(x, y, m, b)

    print(f'm: {m:.3f}, b: {b:.3f}, Loss: {loss:.3f}')
    plot(x,y,m,b)

  return m, b
```

```python
m, b = gradient_descent(x, y, m, b)
```

```python
loss = compute_loss(x, y, m, b)
print(f'loss = {loss:.3f}')
print(f'Learned parameters: m = {m:.3f}, b = {b:.3f}')
```

# linear regression using scikitlearn to learn this model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x_reshaped = x.reshape(-1,1) #this converts a 1D array (a vector) into a 2D array with a single column

print(x)
print(x_reshaped)
print(x.shape)
print(x_reshaped.shape)

model = LinearRegression()
model.fit(x_reshaped, y)
```

```python
m = model.coef_[0]
b = model.intercept_
print(f'learned parameters: m = {m:.3f}, b = {b:.3f}')
```

```python
# Predict y values for training data
y_pred_train = model.predict(x_reshaped)
print(y_pred_train)
print(y)
print(y_pred_train.shape)
print(y.shape)
```

```python
# Compute Mean Squared Error (MSE) Loss
mse = mean_squared_error(y, y_pred_train)
print(f'MSE = {mse:.3f}')
```

```python
x_test = np.arange(0, 2, 0.01).reshape(-1,1)
y_test = model.predict(x_test)
```

```python
plt.plot(x_test, y_test, color='red')
plt.scatter(x, y)
plt.show()
```
